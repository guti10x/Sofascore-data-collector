### GET Clasificación
##### Devuelve la clasificación de todos los equiposen en Laliga.

```python
import requests
import json

API_KEY = '81ff2da84493c8e2bef7f60395c6295c'
API_ENDPOINT = f'https://apiclient.besoccerapps.com/scripts/api/api.php?key={API_KEY}&format=json&req=tables&league=1'

try:
    # Realizar la solicitud GET a la API
    response = requests.get(API_ENDPOINT)

    # Verificar si la solicitud fue exitosa (código de estado 200)
    if response.status_code == 200:
        # Obtener el JSON de la respuesta
        data = response.json()
        
        # Imprimir el JSON de forma legible
        print(json.dumps(data, indent=4))

        # Guardar el JSON en un archivo
        with open('clasificacion_equipos.json', 'w') as file:
            json.dump(data, file, indent=4)

        print('Archivo guardado exitosamente.')
    else:
        # Si la solicitud no fue exitosa, imprimir el código de estado
        print(f'Se produjo un error al hacer la solicitud. Código de estado: {response.status_code}')
except requests.exceptions.RequestException as e:
    # Manejar errores de solicitud
    print(f'Error al hacer la solicitud: {e}')
```

### GET Datos útima jornada disputada
##### Devuelve los datos de todos los partidos de la ultima jornada disputada de Laliga.

```python
import requests
import json

API_KEY = '81ff2da84493c8e2bef7f60395c6295c'
API_ENDPOINT = f'https://apiclient.besoccerapps.com/scripts/api/api.php?key={API_KEY}&format=json&req=matchs&league=1&tz=Europe/Madrid'

try:
    # Realizar la solicitud GET a la API
    response = requests.get(API_ENDPOINT)

    # Verificar si la solicitud fue exitosa (código de estado 200)
    if response.status_code == 200:
        # Obtener el JSON de la respuesta
        data = response.json()
        
        # Imprimir el JSON de forma legible
        print("Datos en formato JSON:")
        print(json.dumps(data, indent=4, ensure_ascii=False))  # La opción ensure_ascii=False permite imprimir caracteres no ASCII correctamente
    else:
        # Si la solicitud no fue exitosa, imprimir el código de estado
        print(f'Se produjo un error al hacer la solicitud. Código de estado: {response.status_code}')
        
    # Guardar el JSON en un archivo
    with open('ultima_jornada_resultados.json', 'w') as file:
        json.dump(data, file, indent=4)
except requests.exceptions.RequestException as e:
    # Manejar errores de solicitud
    print(f'Error al hacer la solicitud: {e}')
```
